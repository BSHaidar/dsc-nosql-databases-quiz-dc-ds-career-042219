## Question 4

What is fault tolerance, and what does it have to do with Resilient Distributed Datasets?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

Fault tolerance is essentially a way to deal with failure, like an unavailable server, a database being unaccessible, and have a way to deal with these unexpected failures to maximize up time. 

Resilient Distribute Datasets have neat way with its built-in fault tolerance approach. For example, if we have 3 databases X, Y, and Z, it would run database X on server1 and store Y and Z databases on server1 for backup, redundancy, and fault tolerance. Similarly, server2 will host Y database and will store X and Z on it. Finally, server3 hosts Z and X, Y are there as backup. At any point a server or a database becomes unaccessible, another server can take its place and pull the backed up database to create a seamless and full-tolerant solution so their user base remain unaffected it.
